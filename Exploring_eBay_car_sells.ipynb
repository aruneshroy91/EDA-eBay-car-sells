# eBay Kleinanzeigen:- A classified section of the German eBay website

* vehicleType - The vehicle Type.
* yearOfRegistration - The year in which the car was first registered.
* gearbox - The transmission type.
* powerPS - The power of the car in PS.

* kilometer - How many kilometers the car has driven.
* monthOfRegistration - The month in which the car was first registered.
* fuelType - What type of fuel the car uses.
* brand - The brand of the car.
* notRepairedDamage - If the car has a damage which is not yet repaired.
* dateCreated - The date on which the eBay listing was created.
* nrOfPictures - The number of pictures in the ad.
* postalCode - The postal code for the location of the vehicle.
* lastSeenOnline - When the crawler saw this ad last online.

```python
autos.rename(columns={'yearOfRegistration':'registration_year',
             'monthOfRegistration':'registration_month',
             'notRepairedDamage':'unrepaired_damage', 
             'dateCreated':'ad_created'},inplace=True)
autos.columns
```

```python
autos.columns[1]
```

```python
def snakejoin(string):
    a = []
    b = [0]
    str_words=[]
    for i in string:
        if i.isupper():
            a.append(string.index(i))
            b.append(string.index(i))
        else:
            pass
    for (i,j) in zip(b,a+[len(string)]):
        str_words.append(string[i:j])
    joined_str_words = ("_".join(str_words)).lower()
    return joined_str_words
```

```python
[snakejoin(col) for col in autos.columns]
```

```python
autos.columns = [snakejoin(col) for col in autos.columns]
```

```python
autos.head()
```

Above we converted the column names from camelcase to snakecase and reworded some of the column names based on the data dictionary to be more descriptive.

* Prices from $50 to $200000 are realistic. Odometer reading from 10000 Km to 200000 Km are realistic. Therefore we remove the outliers.

```python
autos_useful = autos[(autos['odometer_km'].between(10000, 200000)) & (autos['price'].between(50, 200000))]
```

```python
date_crawled_dt = pd.to_datetime(autos_useful['date_crawled']).value_counts(normalize = True, bins = 10, dropna = False).sort_index(ascending = True)*100
last_seen_dt = pd.to_datetime(autos_useful['last_seen']).value_counts(normalize = True, bins = 10, dropna = False).sort_index(ascending = True)*100
ad_created_dt = pd.to_datetime(autos_useful['ad_created']).value_counts(normalize = True, bins = 10, dropna = False).sort_index(ascending = True)*100
```

* Highest and lowest acceptable values for the registration year is 1900 and 2016 respectively. 

```python
car_reg_year_bool = (autos_useful.registration_year > 1900) & (autos_useful.registration_year)<2016
autos_real = autos_useful[car_reg_year_bool]
autos_real.registration_year.value_counts().sort_values(ascending = False)
```

* Most listed cars are bought in 1990 or later.

```python
top_20_brands = autos_real.brand.value_counts(normalize = True).head(20)
top_20_brands
```

* We have taken top 20 brands to aggregate on.

```python
brand_mean_price = {}
for name in top_20_brands.index:
    brand_mean_price[name] = round(autos_real[autos_real.brand == name].price.mean(),2)
bmprice_series = pd.Series(brand_mean_price)
bmprice_df = pd.DataFrame(bmprice_series, columns=['mean price'])
```

* Audi, BMW, Mercedez Benz are most expensive brands.
* Ford and Opel are less expensive
* Volkswagen are in between

```python
brand_mean_ps = {}
for name in top_20_brands.index:
    brand_mean_ps[name] = round(autos_real[autos_real.brand == name].power_p_s.mean(),2)
bmps_series = pd.Series(brand_mean_ps)
bmps_df = pd.DataFrame(bmps_series, columns= ['mean_ps'])
```

```python
bm_df = pd.concat([bmprice_df, bmps_df], axis = 1)
bm_df
```

* Audi, BMW, Mercedez Benz have high powered engines, therefore low milege
* Volkswagen seems to be a better choice due to balanced price and resonable milege, therefore the most bought one.
